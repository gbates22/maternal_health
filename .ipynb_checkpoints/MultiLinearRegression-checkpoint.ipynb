```python
# Perform Multiple Linear Regression

# Here we have 2 variables for multiple regression. If you just want to use one variable for simple linear regression, 
#then use X = df['Interest_Rate'] for example. Alternatively, you may add additional variables within the brackets.
X = df[['avg_body_mass_index','prepregnancy_diabetes_count', 'gestational_diabetes_count','prepregnancy_hypertension_count', 
        'gestational_hypertension_count','hypertension_eclampsia_count']] 
Y = df['total_morbidity_count']
 
# with sklearn
regr = linear_model.LinearRegression()
regr.fit(X, Y)

print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)


# prediction with sklearn
#New_Interest_Rate = 2.75
#New_Unemployment_Rate = 5.3
#print ('Predicted Stock Index Price: \n', regr.predict([[New_Interest_Rate ,New_Unemployment_Rate]]))


# with statsmodels
X = sm.add_constant(X) # adding a constant
 
model = sm.OLS(Y, X).fit()
predictions = model.predict(X) 
 
print_model = model.summary()
print(print_model)
```
