```python
# Data columns for multivariate regression
print(df_rs.columns)
print(df_prm.columns)
```

```python
# Perform Multivariate Regression

scale = StandardScaler()

# Here we have 2 variables for multiple regression. If you just want to use one variable for simple linear regression, 
#then use X = df['Interest_Rate'] for example. Alternatively, you may add additional variables within the brackets.

# No Risk Factors Reported vs. No Morbidity Reported
column_list = ['no_riskfactors_reported']

X = df_rs[column_list] 
Y = df_rs['no_morbidity_reported']

X[column_list] = scale.fit_transform(X[column_list].as_matrix())

print (X)

est = sm.OLS(Y, X).fit()
print(est.summary())

print(est.params)
print(est.bse)
print(est.tvalues)
print(est.pvalues)
```

```python
# Morbidity Events vs. Cesarean Delivery
column_list = ['maternal_transfusion','perineal_laceration', 'ruptured_uterus', 
               'unplanned_hysterectomy','admit_intensive_care']

X = df_rs[column_list] 
Y = df_rs['delivery_cesarean']

X[column_list] = scale.fit_transform(X[column_list].as_matrix())

print (X)

est = sm.OLS(Y, X).fit()
est.summary()
```

```python
# Risk Factors vs. Cesarean Delivery

column_list = ['prepregnancy_diabetes', 'gestational_diabetes',
       'prepregnancy_hypertension', 'gestational_hypertension',
       'hypertension_eclampsia', 'prev_preterm_birth']

X = df_rs[column_list] 
Y = df_rs['delivery_cesarean']

X[column_list] = scale.fit_transform(X[column_list].as_matrix())

print (X)

est = sm.OLS(Y, X).fit()
est.summary()
```
